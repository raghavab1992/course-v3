## Audio to FFT

```python
def wav2img(filename, freq, dest, ymax=1 ,figsize=(8,8)):
    rate, data = wav.read(filename)
    if len(data)>=rate*2:
        data = data[:rate*2]
        data_freq = data[[i for i in range(len(data)) if i%int(rate/freq)==0 ]]
        fft_out = fft(data_freq)
        print(rate, len(data))
        fig, ax = plt.subplots(figsize=figsize)
        ax.plot(np.abs(fft_out)/np.max(np.abs(fft_out)))

        ax.set_axis_off()
        ax.set_ylim([0,ymax])

        plt.savefig(os.path.join(dest,filename.split(".")[0]+".png"),dpi=64,bbox_inches="tight")
        plt.close(fig)
```

```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline
from fastai import *
from fastai.vision import *
```

```python
data = ImageDataBunch.from_csv('/mnt/datadrive1/h216520/fastai_v3/',
                               '/mnt/datadrive1/h216520/fastai_v3/train_audio_image224/',#ds_tfms=get_transforms(),
                              csv_labels = 'labels_subsetn3.csv' ,size=224, bs=32)
data.normalize(imagenet_stats)
```

```python
data.show_batch(rows=3, figsize=(7,6))
```

```python
data.train_dl.one_batch()[0].shape
```

```python
print(data.classes)
len(data.classes),data.c
```

```python
from fastai.vision import models
learn = ConvLearner(data, models.resnet50, metrics=error_rate)
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5, 1e-2)
```

```python
learn.unfreeze()
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5, max_lr=slice(1e-5,5*1e-5))
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, max_lr=slice(1e-6,5*1e-6))
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_top_losses(9, figsize=(15,11))
```
